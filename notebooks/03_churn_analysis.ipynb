- About 25% of the churned customers have an income of 0, along with their number of flights, number of flights with companions, distance in km.

Churned customers exhibit relatively short tenure, with most leaving within the first year of enrollment, indicating early-life churn. While many churned customers show low engagement, a substantial subset demonstrates high flight activity and accumulated value, suggesting that churn is not solely driven by inactivity. The presence of high-CLV churned customers and large volumes of unredeemed points indicates potential dissatisfaction with reward redemption or program value. Overall, churn appears to be driven by both behavioral and structural factors rather than income alone.
