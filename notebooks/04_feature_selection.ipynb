```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/merge.csv')
```

```python
data.head()
```

```python
data[data['tenure'].isnull()]
```

```python
data.info()
```

```python
data = data.drop(['year', 'month', 'day', 'Education', 'cancel_year', 'cancel_month', 'tenure','EnrollmentType','Marital Status','LoyaltyStatus', 'Gender'], axis=1)
```

```python
data.info()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
data.head()
```

```python
X = data.copy()
```

```python
sns.boxplot(x='Income', data=data)
plt.title('Boxplot of Income')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=12)
labels = dbscan.fit_predict(X)

data['dbscan_label'] = labels
```

```python
data['dbscan_label'].value_counts()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(8,6))
plt.scatter(
    X_pca[:,0], X_pca[:,1],
    c=labels, cmap='Set1', alpha=0.6
)
plt.title("DBSCAN Outlier Detection (PCA Projection)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
outliers = data[data['dbscan_label']== -1]
```

```python
outliers.shape
```

```python
outliers.head()
```

```python
outliers.describe()
```

```python
sns.heatmap(
    data.groupby('dbscan_label').mean(),
    cmap='viridis',
    annot=True
)
```

- These outliers are the one with the high CLV. So we can Separate them as they are the important ones.

```python
customer_clean = data[data['dbscan_label'] != -1]
```
